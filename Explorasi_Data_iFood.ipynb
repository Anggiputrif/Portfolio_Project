PREPARATION

```python
#Drop kolom yang memiliki cuma 1 data unik

df = df.drop(['Z_CostContact','Z_Revenue'], axis=1)
```

```python
#Drop kolom yang tidak digunakan

df = df.drop(['Income','Recency','MntGoldProds','NumDealsPurchases','Complain','Response','Customer_Days','marital_Divorced','marital_Married','marital_Single','marital_Together','marital_Widow','education_2n Cycle','education_Basic','education_Graduation','education_Master','education_PhD','MntRegularProds'], axis=1)
```

```python
df.info()
```

```python
#Cek korelasi dari setiap kolom

cor=df.corr()
plt.figure(figsize=(20,15))
sns.heatmap(cor,annot=True,vmin=-1,vmax=1)
plt.title('CORRELATION OF FEATURE')
plt.show()
```

PRODUCT

```python
spent = pd.DataFrame(df.iloc[:,2:7].sum(),columns=['amount_spent']).reset_index()
spent
```

```python
plt.figure(figsize=(11,7))
plt.title("Amount Spent of Products")
yyx= sns.barplot(x='index',y='amount_spent',data=spent);
plt.xlabel('Products')
plt.ylabel('Amount Spent')
for p in yyx.patches:
    yyx.annotate('{:}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
#Menjumlahkan data anak pada kolom baru

df['Kids'] = df['Kidhome'] + df['Teenhome']
```

```python
#Mengelompokkan jumlah produk yang dihabiskan oleh customer yang memiliki anak

expenses = df.groupby(["Kids"])["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts"].mean().unstack().reset_index(name="Total").rename(columns={'level_0': 'expense_type'})
```

```python
#Visualisasi pengelompokan jumlah produk yang dihabiskan oleh customer yang memiliki anak

chart = sns.catplot(data=expenses.sort_values(by="Total",ascending=False),x="expense_type",y="Total",kind="bar",legend=True, col="Kids")
chart.set_xticklabels(rotation=90)
plt.plot()
```

```python
plt.figure(figsize=(13,10),dpi=100)
plt.title('Age Of Customers')
ax = sns.countplot(x=df['Age']);
for p in ax.patches:
    ax.annotate('{:}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
plt.hist(data=df, x="Age")
```

```python
def label(a):
  if a <= 35 :
    return 'Dewasa Awal'
  elif a > 35 and a <= 45 :
    return 'Dewasa Akhir'
  elif a > 45 and a <= 55 :
    return 'Lansia Awal'
  elif a > 55 and a <= 65 :
    return 'Lansia Akhir'
  elif a > 65 :
    return 'Manula'

df['Remarks_Age'] = df['Age'].apply(label)
```

CAMPAIGNS

```python
campaigns = pd.DataFrame(df.iloc[:,11:16].sum(), columns=['amount']).reset_index()
campaigns.head()
```

```python
plt.figure(figsize=(10,5))

plt.title('Success of Each Campaigns')

yx = sns.barplot(x='index',y='amount',data=campaigns);
plt.xlabel('Campaigns')
plt.ylabel('Count')
for p in yx.patches:
    yx.annotate('{:}'.format(p.get_height()), (p.get_x()+0.25, p.get_height()+0.01))
plt.show()
```

```python
grouped = df.groupby(["Remarks_Age"])["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5"].mean().unstack().reset_index(name="Total").rename(columns={'level_0': 'expense_type'})
```

```python
chart2 = sns.catplot(data=grouped.sort_values(by="Total",ascending=False),x="expense_type",y="Total",kind="bar",legend=True, col="Remarks_Age")
chart2.set_xticklabels(rotation=90)
plt.plot()
```
