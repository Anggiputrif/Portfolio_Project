```python
#INISIALISASI MATRIKS PROBABILITAS EMISI
import numpy as np

t = len(T)
v = len(V)
w_given_t = np.zeros((t, v))
w_given_t.shape
```

```python
#PROBABILITAS EMISI
def word_given_tag(word, tag, train_bag = cleaned_train_set):
    tag_list = [pair for pair in train_bag if pair[1]==tag]
    count_tag = len(tag_list)
    w_given_tag_list = [pair[0] for pair in tag_list if pair[0]==word]
    count_w_given_tag = len(w_given_tag_list)
    return (count_w_given_tag, count_tag)
```

```python
#PROBABILITAS TRANSISI
def t2_given_t1(t2, t1, train_bag = cleaned_train_set):
    tags = [pair[1] for pair in train_bag]
    count_t1 = len([t for t in tags if t==t1])
    count_t2_t1 = 0
    for index in range(len(tags)-1):
        if tags[index]==t1 and tags[index+1] == t2:
            count_t2_t1 += 1
    return (count_t2_t1, count_t1 ,(count_t2_t1/count_t1)*100)
```

```python
#MATRIKS PROBABILITAS TRANSISI
tags_matrix = np.zeros((len(T), len(T)), dtype='float32')
for i, t1 in enumerate(list(T)):
  for j, t2 in enumerate(list(T)):
    tags_matrix[i, j] = t2_given_t1(t2, t1)[0]/t2_given_t1(t2, t1)[1]
```

```python
import pandas as pd

tags_df = pd.DataFrame(tags_matrix, columns = list(T), index=list(T))
tags_df
```

```python
def Viterbi(words, train_bag=cleaned_train_set):
    state = []
    T = list(set([pair[1] for pair in train_bag]))

    for key, word in enumerate(words):
        p = []
        for tag in T:
            if key == 0:
                try:
                    transition_p = tags_df.loc['.', tag]
                except KeyError:
                    transition_p = 1/len(T)
            else:
                try:
                    transition_p = tags_df.loc[state[-1], tag]
                except KeyError:
                    transition_p = 1/len(T)

            emission_p = word_given_tag(word, tag)[0] / word_given_tag(word, tag)[1]

            state_probability = emission_p * transition_p
            p.append(state_probability)

        pmax = max(p)
        state_max = T[p.index(pmax)]
        state.append(state_max)
    return list(zip(words, state))
```

```python
# Buat daftar semua kata benda unik dalam transaksi
all_words = list(set(word for transaction in transactions_with_word for word in transaction))

# One-hot encode transaksi
one_hot_encoded_transactions = []
for transaction in transactions_with_word:
    one_hot_encoded_transactions.append([1 if word in transaction else 0 for word in all_words])
 
df_transactions = pd.DataFrame(one_hot_encoded_transactions, columns=all_words)
print(df_transactions.head()) 
```

```python
from itertools import combinations

# Menghitung frequent itemsets secara manual
def get_frequent_itemsets(df, min_support):
    itemsets = []
    support_counts = Counter()

    for row in df.iterrows():
        row_items = [item for item, present in zip(df.columns, row[1]) if present]
        for size in range(1, len(row_items) + 1):
            for combination in combinations(row_items, size):
                support_counts[combination] += 1

    for itemset, count in support_counts.items():
        support = count / len(df)
        if support >= min_support:
            itemsets.append((itemset, support))

    return pd.DataFrame(itemsets, columns=['itemsets', 'support'])

min_support = 0.1  # Sesuaikan dengan kebutuhan
frequent_itemsets = get_frequent_itemsets(df_transactions, min_support)
print(frequent_itemsets)
```

```python
min_nilai_support = frequent_itemsets['support'].min()
print("min_support:", min_nilai_support)
```

```python
# Menghitung aturan asosiatif secara manual
def get_association_rules(frequent_itemsets, min_confidence):
    rules = []
    for itemset, support in frequent_itemsets.itertuples(index=False):
        if len(itemset) > 1:
            for item in itemset:
                antecedent = tuple(sorted(set(itemset) - {item}))
                antecedent_support = frequent_itemsets[frequent_itemsets['itemsets'].apply(lambda x: set(x) == set(antecedent))]['support'].values
                if len(antecedent_support) > 0:
                    confidence = support / antecedent_support[0]
                    if confidence >= min_confidence:
                        rules.append((antecedent, item, confidence, support))
    return pd.DataFrame(rules, columns=['antecedent', 'consequent', 'confidence', 'support'])

min_confidence = 0.1 # Sesuaikan dengan kebutuhan
rules = get_association_rules(frequent_itemsets, min_confidence)

rules_with_word = rules[rules['antecedent'].apply(lambda x: 'aplikasi' in x)]
print(rules_with_word)
```
